```python
%matplotlib inline

import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import random
import math
```

```python
def graficiSceltaBK(title, fileURL):
  intervals1 = []
  sample_means1 = []
  x = []
  y_gurobi = []
  y_bnb = []
  url = fileURL

  df = pd.read_csv(url)
  df = df.sort_values(by=['jobs'])
  df.drop_duplicates(subset=['jobs'])

  for i in df.index:
    x.append(df.loc[i,"jobs"])
    y_gurobi.append(df.loc[i,"gurobiTime"])
    y_bnb.append(df.loc[i,"bnbTime"])
  
  # plotting graph
  fig, ax = plt.subplots(1,figsize=(20,7.5))
  fig.suptitle('Vertically stacked subplots')

  ax.plot(x, y_gurobi, label = "gurobi")
  ax.errorbar(x, y_gurobi, fmt ='o')
  ax.set_title(title)
  ax.set_xlabel('jobs')
  ax.set_ylabel('sec')


  ax.plot(x, y_bnb, label = "branch & bound",color = 'y')
  ax.errorbar(x, y_bnb, fmt ='o')
  #ax.axhline(y = 60)

  plt.legend(loc="upper left")

  #ax.set_title('k = 64')
```

```python
graficiSceltaBK('Execution time', 'https://raw.githubusercontent.com/GiulioAppetito/AMOD_Project_2021_22/master/results/few%20close%20to%20zero.csv')
```
